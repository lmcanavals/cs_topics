```python
import random
import string
import re
import os
```

```python
corpus_text = 'vallejo.txt'
text = None

with open(corpus_text, 'r') as f:
    text = f.read()

words_order = [s.lower() for s in re.compile('\w+').findall(text)]
words = list(set(words_order))

print("random sentence of size 10")
print(' '.join(random.choices(words, k=10)))
```

```python
rhyme_groups = {}
for w in words:
    if len(w) < 4:
        continue

    end_of_word = w[-2:]
    if w[-3] in 'aeiou' or w[-2] in 'aeiou':
        end_of_word = w[-3] + end_of_word
    elif w[-2] in 'rl':
        end_of_word = '*' + end_of_word
    else:
        end_of_word = w[-3] + end_of_word
    if end_of_word in rhyme_groups:
        rhyme_groups[end_of_word].append(w)
    else:
        rhyme_groups[end_of_word] = [w]

for k in rhyme_groups.keys():
    print(rhyme_groups[k])
```

```python
key_to_del = []
for key in rhyme_groups:
    value = rhyme_groups[key]
    if len(value) < 3:
        key_to_del.append(key)
for key in key_to_del:
    del rhyme_groups[key]
for k in rhyme_groups.keys():
    print(k, rhyme_groups[k])
```

```python
n_random_words = 4
n_sentences = 2

sequence = []
for _ in range(n_sentences):
    rhyme = random.sample([*rhyme_groups], 1)[0]
    for _ in range(2):
        for _ in range(n_random_words):
            sequence += random.sample(words, 1)
        sequence += random.sample(rhyme_groups[rhyme], 1)

for i, w in enumerate(sequence):
    print(f"{w} ", end="" if i % (n_random_words + 1) != n_random_words else "\n")
```

```python
transition = {}
for i, w in enumerate(words_order[:-1]):
    if w in transition:
        transition[w].append(words_order[i + 1])
    else:
        transition[w] = [words_order[i + 1]]

transition = {word: list(set(nexts)) for word, nexts in transition.items()}

for v in transition:
    print(v, transition[v])
```

```python
sequence_length = 10

sequence = random.sample(words, 1)
for i in range(sequence_length - 1):
    sequence += random.sample(transition[sequence[i]], 1)
print(' '.join(sequence))
```
