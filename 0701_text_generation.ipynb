```python
import random
import string
import re
import os
```

```python
corpus_text = 'vallejo.txt'
text = None

with open(corpus_text, 'r') as f:
    text = f.read()

words_order = [s.lower() for s in re.compile('\w+').findall(text)]
words = list(set(words_order))

print("random sentence of size 10")
print(' '.join(random.choices(words, k=10)))
```

```python
rhyme_groups = {}
for w in words:
    if len(w) < 4:
        continue

    end_of_word = w[-2:]
    if w[-3] in 'aeiou' or w[-2] in 'aeiou':
        end_of_word = w[-3] + end_of_word
    elif w[-2] in 'rl':
        end_of_word = '*' + end_of_word
    else:
        end_of_word = w[-3] + end_of_word
    if end_of_word in rhyme_groups:
        rhyme_groups[end_of_word].append(w)
    else:
        rhyme_groups[end_of_word] = [w]

for k in rhyme_groups.keys():
    print(rhyme_groups[k])
```

```python
key_to_del = []
for key in rhyme_groups:
    value = rhyme_groups[key]
    if len(value) < 3:
        key_to_del.append(key)
for key in key_to_del:
    del rhyme_groups[key]
for k in rhyme_groups.keys():
    print(k, rhyme_groups[k])
```

```python
n_random_words = 4
n_sentences = 2

sequence = []
for _ in range(n_sentences):
    rhyme = random.sample([*rhyme_groups], 1)[0]
    for _ in range(2):
        for _ in range(n_random_words):
            sequence += random.sample(words, 1)
        sequence += random.sample(rhyme_groups[rhyme], 1)

for i, w in enumerate(sequence):
    print(f"{w} ", end="" if i % (n_random_words + 1) != n_random_words else "\n")
```

```python
transition = {}
for i, w in enumerate(words_order[:-1]):
    if w in transition:
        transition[w].append(words_order[i + 1])
    else:
        transition[w] = [words_order[i + 1]]

transition = {word: list(set(nexts)) for word, nexts in transition.items()}

for v in transition:
    print(v, transition[v])
```

```python
sequence_length = 10

sequence = random.sample(words, 1)
for i in range(sequence_length - 1):
    sequence += random.sample(transition[sequence[i]], 1)
print(' '.join(sequence))
```

```python
from ortools.sat.python import cp_model
```

```python
sequence_length = 20
min_dom = 0
max_dom = len(words)-1

model = cp_model.CpModel()

seq = [model.NewIntVar(min_dom, max_dom, f'seq{i}') for i in range(sequence_length)]

word2i = {word: i for i, word in enumerate(words)}
transition_table = []

for w1 in transition:
    for w2 in transition[w1]:
        transition_table.append([word2i[w1], word2i[w2]])

for i, v in enumerate(seq[:-1]):
    model.AddAllowedAssignments( # table constrain
        (seq[i], seq[i+1]), # variables
        transition_table # table
    )

rhymes_table = []
for l in rhyme_groups.values():
    rhymes_table += [[word2i[w1], word2i[w2]] for w1 in l for w2 in l]
model.AddAllowedAssignments((seq[4], seq[9]), rhymes_table)
model.AddAllowedAssignments((seq[14], seq[19]), rhymes_table)

plagiarism_table = [[word2i[words_order[i]], word2i[words_order[i+1]],
                     word2i[words_order[i+2]], word2i[words_order[i+3]]]
                    for i, w in enumerate(words_order[:-3])]

for i, w in enumerate(seq[:-3]):
    model.AddForbiddenAssignments((seq[i], seq[i+1], seq[i+2], seq[i+3]),
                                  plagiarism_table)

model.AddAllDifferent(seq)

solver = cp_model.CpSolver()
status = solver.Solve(model)
```

```python
for i, w in enumerate(seq):
    print(f"{words[solver.Value(w)]} ", end="" if i % 5 != 4 else "\n")

with open('output.txt', 'w') as f:
    s = ""
    for i, w in enumerate(seq):
        s += f"{words[solver.Value(w)]} "
        if  i % 5 == 4:
            f.write(s)
            f.write('\n')
            s = ''
    f.close()
```

```python
class PoemSolutionPrinter(cp_model.CpSolverSolutionCallback):
    def __init__(self, variables):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.__variables = variables
        self.__solution_count = 0

    def on_solution_callback(self):
        self.__solution_count += 1

        for i, w in enumerate(self.__variables):
            print(f"{words[self.Value(w)]} ", end="" if i % 5 != 4 else "\n")
        print()

    def solution_count(self):
        return self.__solution_count

solver = cp_model.CpSolver()
solution_printer = PoemSolutionPrinter(seq)
solver.parameters.enumerate_all_solutions = True
status = solver.Solve(model, solution_printer)

print(f"Status = {solver.StatusName(status)}")
print(f"Number of solutions found = {solution_printer.solution_count()}")
```
